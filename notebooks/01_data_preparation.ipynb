```python
file_name = 'historical_data.csv'
```

```python
df_hist = pd.read_csv(folder_name + file_name)
```

```python
df_hist.head()
```

```python
df_hist.describe()
```

```python
# Rename columns to snake_case (professional standard)
df_hist.columns = df_hist.columns.str.strip().str.replace(' ','_').str.lower()
```

```python
#keeping only useful column droping noise
df_hist = df_hist[['account','coin','size_usd','side','timestamp_ist','direction','closed_pnl','fee']]
```

```python
df_hist.head()
```

```python
# convert to datetime first
df_hist['timestamp_ist'] = pd.to_datetime(df_hist['timestamp_ist'], dayfirst=True)
```

```python
# now extract date
df_hist['date'] = df_hist['timestamp_ist'].dt.normalize()
```

```python
# Step 3 - check
print(df_hist['direction'].value_counts())
print(df_hist[df_hist['closed_pnl'] != 0].shape[0])
```

```python
#keep only rows where actual PnL was realized
df_closed = df_hist[df_hist['closed_pnl']!=0].copy()
```

```python
df_closed.head()
```

```python
print(df_closed.shape)
print(df_closed['direction'].value_counts())
```

```python
#keeps only voluntary closed trades

keep_directions = ['Close Long','Close Short','Sell']
```

```python
df_closed = df_closed[df_closed['direction'].isin(keep_directions)].copy()
```

```python
print(df_closed['direction'].value_counts())
```

```python
# Merge trader data with sentiment data

df_merged = pd.merge(df_closed, df_index[['date','value','classification','sentiment_binary']],
                     on='date',
                     how='inner')
```

```python
df_merged.head()
```

```python
df_merged.info()
```

```python
#drop one timestamp came when merging happened
df_merged = df_merged.drop(columns=['timestammp_ist'])
```

```python
print(df_merged.shape)
print(df_merged.columns.tolist())
```

```python
print(df_merged[['timestamp_ist', 'date']].head())
```

```python
#group by each trader total trades per day

daily = df_merged.groupby(['date','account']).agg(
    daily_pnl = ('closed_pnl','sum'),
    trade_count = ('closed_pnl','count'),
    avg_size_usd = ('size_usd','mean'),
    total_fees = ('fee','sum'),
    win_rate = ('closed_pnl',lambda x : (x > 0).mean()),
    long_ratio = ('side',lambda x : (x.str.upper() =='BUY').mean())
).reset_index()
```

```python
daily['net_pnl'] = daily['daily_pnl'] - daily['total_fees']
```

```python
daily = daily.merge(
    df_merged[['date','value','classification','sentiment_binary']].drop_duplicates('date'),
    on = 'date',
    how = 'left'
)
```
